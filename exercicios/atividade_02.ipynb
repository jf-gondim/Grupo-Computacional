```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivada(f, a, metodo, h):
    if metodo == 'central':
        return (f(a + h) - f(a - h))/(2*h)
    elif metodo == 'forward':
        return (f(a + h) - f(a))/h
    elif metodo == 'backward':
        return (f(a) - f(a - h))/h
    else:
        raise ValueError("Método precisa ser do tipo 'central', 'forward' ou 'backward'.")
```

# Questão 1

```python
x = np.linspace(-5,5,200)
f = lambda x: ((7*x**3 - 5*x +1) / (2*x**4 + x**2 + 1))
y = f(x)
dydx = derivada(f, x, 'central', 0.00001)

fun = lambda x: ((-14*x**6 + 37*x**4 - 8*x**3 + 26*x**2 - 2*x -5) / ((2*x**4 + x**2 + 1)**2))
y_fun = fun(x)
```

```python
f_exato = lambda x: ((-14*x**6 + 37*x**4 - 8*x**3 + 26*x**2 - 2*x -5) / ((2*x**4 + x**2 + 1)**2))
y_exato = f_exato(x)
```

```python
plt.figure(figsize=(12,5))
plt.rcParams.update({'font.size': 14})
plt.plot(x, y, label='y=f(x)')
plt.plot(x, dydx, label="Diferença Central | y=f'(x)")
plt.plot(x, y_fun, '--', label="Resultado Analítico | y=f'(x)")
plt.legend()
plt.grid(True)
plt.show()
```

# Questão 2

Para calcular o polinômio de quarto grau $T_{4}(x)$ centrado em $x=0$ para a função $f(x)=\cos(x)+\sin(2x)$ no intervalo $[-\pi,\pi]$. Plotamos primeiro o gráfico $y=f(x)$

```python
x = np.linspace(-np.pi, np.pi, 200)
f = lambda x: np.cos(x)+np.sin(2*x)
y = f(x)

plt.plot(x, y)
plt.show()
```

Vamos agora calcular os coeficientes $a_{n}=\frac{f^{(n)}(0)}{n!}$ para $n=0,1,2,3,4$:

```python
from scipy.misc import derivative
```

```python
a0 = f(0)
a1 = derivative(f, 0, dx=0.001, n=1)
a2 = derivative(f, 0, dx=0.001, n=2) / 2
a3 = derivative(f, 0, dx=0.001, n=3, order=5) / 6
a4 = derivative(f, 0, dx=0.001, n=4, order=5) / 24
```

```python
print(a0, a1, a2, a3, a4)
```

```python
8/6
```

```python
T4 = a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4
```

```python
plt.rcParams.update({'font.size': 14})
plt.plot(x, y, label=r"$f(x)$")
plt.plot(x, T4, label=r"$T_{4}(x)$")
plt.xlim([-np.pi,np.pi])
plt.ylim([-np.pi,np.pi])
plt.legend(loc="best")
plt.show()
```
