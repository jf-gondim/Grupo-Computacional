# Tópico 4 (opcional)

```python
import random as rd
```

```python
def func_teste(x):
    return np.sin(x)

def roll_int():
    return rd.randint(0,1)
```

```python
x4 = np.linspace(0, 2*np.pi, 100)
num_aleatorios = []
matriz_final = []
```

```python
for i in range(len(x4)):
    n_aleatorio = rd.random() * 0.4
    num_aleatorios.append(n_aleatorio)
```

```python
for i in range(len(x4)):
    k = roll_int()
    if (k == 0):
        matriz_final.append(func_teste(x4[i]) + num_aleatorios[i])
    else:
        matriz_final.append(func_teste(x4[i]) - num_aleatorios[i])
```

```python
plt.figure(figsize=(8,6))
plt.plot(x4, matriz_final, 'ko', label='Dados aleatórios')
plt.plot(x4, np.sin(x4), label='Valor exato')
plt.title("Gráfico 4")
plt.xlabel("x")
plt.ylabel("y")
#plt.xlim(0,np.pi)
plt.legend()
plt.show()
```
